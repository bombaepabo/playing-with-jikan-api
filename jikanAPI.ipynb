```python
def topanime_upcoming(n):
  rank= [ ]
  title=[]
  url= []
  for i in range(1,n+1):
    top_anime_upcoming = jikan.top(type='anime',page=i,subtype='upcoming')
    for i in top_anime_upcoming['top']:
      rank.append(i['rank'])
      title.append(i['title'])
      url.append(i['url'])
  mydf = pd.DataFrame(list(zip(rank,title,url)), columns = ['rank', 'title','urls'])

  return mydf
b = topanime_upcoming(2)
b
```

```python
def anime_later():
  title=[]
  image=[]
  later_season = jikan.season_later()
  for i in later_season['anime']:
    title.append(i['title'])
    image.append(i['url'])
  mydf = pd.DataFrame(list(zip(title,image)), columns = ['title', 'url'])

  return mydf 
c = anime_later()
c.to_csv('anime_upcoming.csv', encoding='utf-8')
c.head(500)
```

```python
#search anime
def search_anime(name) :
  title = []
  members = []
  score = []
  typed = []
  episode = []
  for j in range(1,4):
    anime_ep = jikan.search('anime',name,page=j)

    for i in anime_ep['results']:
      title.append(i['title'])
      members.append(i['members'])
      score.append(i['score'])
      typed.append(i['type'])
      episode.append(i['episodes'])
  mydf = pd.DataFrame(list(zip(score,title,members,episode,typed)), columns = ['rating', 'title','members','episode','type'])
  return mydf
search_= search_anime('boku no hero',)
search_.to_csv('search_anime.csv', encoding='utf-8')
```

```python
#Anime search by Studio 
def anime_by_studio(id,n):
    title=[]
    score=[]
    company=[]
    episode=[]
    member=[]
    for i in range(1,n+1):
      produce_search = jikan.producer(producer_id=id,page=i)
    
      for i in produce_search['anime']:
        title.append(i['title'])
        score.append(i['score'])
        episode.append(i['episodes'])
        member.append(i['members'])
        for j in i['producers']:
          company.append(j['name'])
    mydf = pd.DataFrame(list(zip(company,score,title,episode,member)), columns = ['company', 'rating','title','ep','member'])
    return mydf
    
studio_anime = anime_by_studio(21,1)
studio_anime.to_csv('Anime_Ghibli.csv', encoding='utf-8')
studio_anime.head(500)
```

```python
def top_manga(n):
  title=[]
  rating = []
  members = []
  for i in range(1,n+1):
    top_manga = jikan.top(type='manga',page=n)
    for i in top_manga['top']:
      title.append(i['title'])
      rating.append(i['score'])
      members.append(i['members'])
  #df = pd.DataFrame({"title:",title,"rating",rating,"members",members})
  mydf = pd.DataFrame(list(zip(title,rating,members)), columns = ['title', 'rating','members'])
  return mydf 
topmanga = top_manga(20)
topmanga.to_csv('top_manga.csv', encoding='utf-8')    
```

```python
#เอามาลอง groupby ดู แยกแต่ละปี ของ genre เดียว
def anime_by_genre(genre):
  anime_popular_bygenre = jikan.genre('anime',genre_id=genre)
  members = []
  score = []
  title = []
  year = []
  for i in anime_popular_bygenre['anime']:
    members.append(i['members'])
    score.append(i['score'])
    title.append(i['title'])
    year.append(i['airing_start'])
    
  mydf = pd.DataFrame(list(zip(title,score,members,year)), columns = ['title', 'score','members','airing start'])
  only_year = []
  for i in mydf['airing start']:
    if i == None:
      only_year.append('older years')
    else:
      only_year.append(i[0:4])
    
  mydf['only_year'] = only_year
  mydf = mydf.groupby(['only_year']).agg({'members':'sum','score':'mean'})
  return mydf  


choice =""" 1 = Action 
            2 = Adventure 
            3 = Cars 
            4 = Comedy 
            5 = Dementia 
            6 = Demons 
            8 = Drama 
            9 = Ecchi 
            10 = Fantasy 
                                    11 = Game 
                                    35 = Harem 
                                    12 = Hentai 
                                    13 = Historical 
                                    14 = Horror 
                                    43 = Josei 
                                    15 = Kids 
                                    16 = Magic 
                                    17 = Martial Arts 
                                    18 = Mecha 
                                    38 = Military 
            19 = Music 
            7 = Mystery 
            20 = Parody 
            39 = Police 
            40 = Psychological 
            22 = Romance 
            21 = Samurai 
            23 = School 
            24 = Sci-Fi 
                                    42 = Seinen 
                                    25 = Shoujo 
                                    26 = Shoujo Ai 
                                    27 = Shounen 
                                    28 = Shounen Ai 
                                    36 = Slice of Life 
                                    29 = Space 
                                    30 = Sports 
                                    31 = Super Power 
                                    37 = Supernatural 
                                    41 = Thriller 
                                    32 = Vampire 
                                    33 = Yaoi 
                                    34 = Yuri 
"""
df = anime_by_genre(42)
df.to_csv('anime_by_genre.csv', encoding='utf-8')
df.head(100)
```

```python
def top_anime_byspecifictime(startyear,endyear,n):
  title = []
  score = []
  member = []
  year = []
  for j in range(n):
    anime_popular_bygenre = jikan.top(type='anime',page=j)
    for i in anime_popular_bygenre['top']:
      year_specific = int(i['start_date'][4:])
      if year_specific >= startyear and year_specific <= endyear :
        title.append(i['title'])
        score.append(i['score'])
        member.append(i['members'])
        year.append(i['start_date'][4:])
  mydf = pd.DataFrame(list(zip(title,score,member,year)), columns = ['title', 'score','members','years'])
  return mydf






anime = top_anime_byspecifictime(2000,2010,6)
anime.to_csv('top_anime_2010-2020.csv', encoding='utf-8')
anime.head(50)
```
